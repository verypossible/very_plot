```python
import sys
sys.path.append("../")
import numpy as np
import matplotlib.pyplot as plt
import functools

from very_plot import themes
themes.blog_mpl()

%matplotlib inline
```

## Time Series Plot

```python
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111)

x = np.arange(0, 10, 0.1)

for i in range(4):
    ax.plot(x, np.sin(2*x + i), label=i)

ax.legend(fontsize=16)
ax.despine()
```

## Histogram

```python
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111)


for loc in np.arange(0, 35, 5): 
    x = np.random.normal(loc, 1, 100)
    ax.hist(x)

ax.despine()
```

## Scatter

```python
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111)


for loc in np.arange(0, 50, 5): 
    x = np.random.normal(loc, 1, 20)
    y = np.random.normal(loc - np.random.random() * loc, 1, 20)
    ax.scatter(x, y, s=50, label=loc)

ax.despine()
ax.legend(fontsize=16);
```

## Grouped Bars

```python
data = np.random.rand(5,5)*5+2

x = np.arange(data.shape[0])
dx = (np.arange(data.shape[1])-data.shape[1]/5.)/(data.shape[1]+5.)
d = 1./(data.shape[1]+5.)


fig, ax=plt.subplots(figsize=(12, 9))
for i in range(data.shape[1]):
    ax.bar(x+dx[i],data[:,i], width=d, label=f"label {i}")

ax.set_ylim(0,10)
ax.legend(framealpha=1, fontsize=16);
```
